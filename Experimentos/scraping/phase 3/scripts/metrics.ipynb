```python
path_errors = "../logs/errors.txt"

with open(path_errors, 'r') as f:
    errors = f.readlines()
```

```python
# extraindo os erros e quantas vezes eles ocorreram
pattern = r'Error:\s*(.+)'

erros = re.findall(pattern, ''.join(errors))

erros 
```

```python
string = "HTTPSConnectionPool(host='web.archive.org', port=443): Max retries exceeded with url:"

# substituindo o conteúdo do erro pela string genérica
erros = [string if string in erro else erro for erro in erros]
```

```python
for erro in erros:
    if erro == string:
        print(erro)
```

```python
# contando a quantidade de vezes que cada erro ocorreu
erros_count = pd.Series(erros).value_counts()
print(erros_count)
```
